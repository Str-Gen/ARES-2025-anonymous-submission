```python
from fastai.tabular.all import df_shrink
```

```python
df_train = pd.read_parquet('/kaggle/input/ember-2017-v2-features/train_ember_2017_v2_features.parquet')
df_test = pd.read_parquet('/kaggle/input/ember-2017-v2-features/test_ember_2017_v2_features.parquet')
```

```python
df_train.shape, df_test.shape
```

```python
df_train.replace([np.inf, -np.inf], np.nan, inplace=True)    
print(df_train.isna().any(axis=1).sum(), "rows with at least one NaN to remove")
df_train.dropna(inplace=True)
```

```python
df_test.replace([np.inf, -np.inf], np.nan, inplace=True)    
print(df_test.isna().any(axis=1).sum(), "rows with at least one NaN to remove")
df_test.dropna(inplace=True)
```

```python
print(df_train.duplicated().sum(), "fully duplicate rows to remove")
df_train.drop_duplicates(inplace=True)
df_train.reset_index(inplace=True, drop=True)
print(df_train.shape)
```

```python
print(df_test.duplicated().sum(), "fully duplicate rows to remove")
df_test.drop_duplicates(inplace=True)
df_test.reset_index(inplace=True, drop=True)
print(df_test.shape)
```

# Zero variance columns?


```python
zvr = df_train.nunique() != 1
zvr.index[zvr == False], len(zvr.index[zvr == False])
```

```python
zvr = df_test.nunique() != 1
zvr.index[zvr == False], len(zvr.index[zvr == False])
```

```python
df_train = df_shrink(df_train, obj2cat=False, int2uint=False)
df_test = df_shrink(df_test, obj2cat=False, int2uint=False)
df_train.to_parquet('train_ember_2017_v2_features.parquet')
df_test.to_parquet('test_ember_2017_v2_features.parquet')
```
